```python

# Step 1: Get the unique years and sort them in ascending order
unique_years = sorted(df['year'].unique(), reverse=False)

# Step 2: Create a mapping from the highest unique year (which should remain 2024) to descending years
year_mapping = {year: 2024 - i for i, year in enumerate(unique_years)}

# Step 3: Replace the years in the DataFrame with the new mapping
df['year'] = df['year'].map(year_mapping)

# Step 4: Verify the results
print(df['year'].unique())  # This should now show the mapped years starting from 2024 and descending
```

```python
df.head()
```

```python
# Step 1: Convert 'date_of_journey' to datetime 
df['Date of Journey'] = pd.to_datetime(df['Date of Journey'], errors='coerce')

# Step 2: Extract the year from 'date_of_journey'
df['journey_year'] = df['Date of Journey'].dt.year

# Step 3: Map the years in 'journey_year' using the 'year_mapping' dictionary
df['journey_year_mapped'] = df['journey_year'].map(year_mapping)

# Step 4: Replace the year in 'date_of_journey' with the mapped year only if 'journey_year_mapped' is not NaN
df['Date of Journey'] = df.apply(
    lambda x: x['Date of Journey'].replace(year=int(x['journey_year_mapped'])) 
    if pd.notnull(x['journey_year_mapped']) else x['Date of Journey'], 
    axis=1
)

# Step 5: Drop the temporary 'journey_year' and 'journey_year_mapped' columns
df.drop(columns=['journey_year', 'journey_year_mapped'], inplace=True)

# Step 6: Verify the changes
print(df['Date of Journey'].head())

```

# Dividing into Training and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(columns='Confirmation Status') #these will be our features
y=df['Confirmation Status'] #this is our target variable
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

# Applying PCA

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.fit_transform(X_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
X_train = pca.fit_transform(X_train_scaled)
X_test= pca.fit_transform(X_test_scaled)
```

#  Lets use different models and check which will be the best for prediction

# 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,f1_score
```

```python
# Accuracy on training data
X_train_prediction=model.predict(X_train)
accuracy_score(X_train_prediction,y_train)
```

```python
# Accuracy on testing data
X_test_prediction=model.predict(X_test)
accuracy_score(X_test_prediction,y_test)
```

```python
from sklearn.metrics import classification_report, roc_auc_score
print(classification_report(y_test, y_pred))
print(roc_auc_score(y_test, y_pred))
```
